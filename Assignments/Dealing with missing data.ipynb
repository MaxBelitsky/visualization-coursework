```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', -1)  # or 199
```

```python
df = pd.read_excel('dataset.xlsx')
df.shape
```

```python
p = df.isnull().sum()
p
```

```python
thresh=0.8

a = df.loc[:,p/df.shape[0] < thresh]
b = a.dropna(thresh=a.shape[1])
b.shape
```

```python
b.head()
```

```python
b.to_excel('cleaned_thresh_' + str(thresh) +'.xlsx')
```
