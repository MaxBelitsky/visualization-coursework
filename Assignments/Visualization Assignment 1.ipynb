```python
import pandas as pd
import numpy as np
```

```python
def parsing_excel(filename):
    return pd.read_excel(filename)

df = parsing_excel('../dataset.xlsx')
df.head()
```

```python
def freq_age(df):
    return df['Patient age quantile'].value_counts()

freq_age(df)
```

```python
def nan_attributes(df):
    return df.isnull().sum(axis=0)

nan_attributes(df)
```

```python
df1 = nan_attributes(df)
df1.to_csv("nan_attributes.txt", sep=' ', index=True, header=False)
```

```python
def nan_individual(df):
    return df.isnull().sum(axis=1)

missingrow = nan_individual(df)
nan_individual(df)
```

```python
df2 = nan_individual(df)
df2.to_csv("nan_individual.txt", sep=' ', index=True, header=False)
```
